```python
import pandas as pd 
import sqlite3
```

```python

df = pd.read_csv('Data_Wrangling.csv')
df.drop(columns=['title', 'Price VND', 'Unnamed: 11', 'Unnamed: 12'],inplace=True)
df.columns = df.columns.str.strip()
df
```

```python
conn = sqlite3.connect('real_estate.db')
```

```python
df.to_sql('reestate_infor', conn, if_exists='replace')
conn.close()
```

```python
def printAll(listOfResults):
    for i in listOfResults:
        print(i)

conn = sqlite3.connect('real_estate.db')

c = conn.cursor()

print("\n Checking the amount of category:")
c.execute("SELECT e.Type,COUNT(*) FROM reestate_infor e GROUP BY e.Type")
printAll(c.fetchall())


print("\n Average price by type:" )
c.execute("SELECT e.Type, AVG(Price_USD) FROM reestate_infor e GROUP BY e.Type")
printAll(c.fetchall())


print("\n Number of Properties by Location:")
c.execute("SELECT e.district, COUNT(*) FROM reestate_infor e GROUP BY e.district")
printAll(c.fetchall())

conn.commit()
conn.close()
```
